```python
import pandas as pd
import numpy as np
import os
import glob
```

```python
def make_prompts(concepts='a person, a car, park, beach',num_prompts_per_concept=5,name=None):
    name_ = concepts.replace('a ','').replace(' ','_').replace(',','')
    if name is not None:
        name_ = name
    name_ = f'prompts/prompts-{name_}.csv'
    concepts = concepts.split(',')
    concepts = [c.strip() for c in concepts]
    
    prompts_ = []
    concepts_ = []
    
    for concept in concepts:
        prompts_.extend([f'image of {concept}']*num_prompts_per_concept)
        prompts_.extend([f'picture of {concept}']*num_prompts_per_concept)
        prompts_.extend([f'photo of {concept}']*num_prompts_per_concept)
        prompts_.extend([f'portrait of {concept}']*num_prompts_per_concept)
        prompts_.extend([f'{concept}']*num_prompts_per_concept)

        concepts_.extend([concept]*5*num_prompts_per_concept)
    seeds_  = np.random.randint(0,2**16,(len(prompts_),))
    
    case_numbers = np.arange(len(prompts_))
    
    df = pd.DataFrame({'case_number': case_numbers, 'prompt':prompts_, 'evaluation_seed':seeds_,'concept':concepts_})
    
    df.to_csv(name_,index=False)
    return df
```

```python
make_prompts(concepts='a person, man, woman, doctor, barista',num_prompts_per_concept=20,name='person')
```

```python
make_prompts(concepts='food, a steak, a pizza, salad, fruit bowl',num_prompts_per_concept=20,name='food')
```
